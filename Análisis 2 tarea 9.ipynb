```python
import math
import random
import matplotlib.pyplot as plt
```

```python
def T(x):
    if x != 0:
        t = 1/x - math.floor(1/x) 
    else:
        t = 0
    return t

            
```

```python
L = [0]
for i in range(1000):
    l = random.random()
    L.append(l)

    
```

```python
L2 = []
for i in range(len(L)):
    s = T(L[i])
    L2.append(s)
    
```

```python
#print(L,L2)
```

```python
plt.scatter(L,L2,marker = '.', color='black')
```
